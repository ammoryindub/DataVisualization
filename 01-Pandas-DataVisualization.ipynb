# DataVisualization using Pandas

We can modify the sample data or add our own dataset to the DataFrame to visualize our data. Additionally, you can customize the plot titles, labels, colors, and other properties as per your requirements 


```python
# Import libraries

import numpy as np 

import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
import seaborn as sns
```

```python
# we call our dataset

df = pd.read_csv('dataset.csv')
```

```python
df
```

#### let's start with performing a scatter plot using DataFrames, to visualize "column a" against "column b"

```python
# calling plot on dataframe

df.plot.scatter(x='a', y='b')
```

```python
# using scatter plot arguments

df.plot.scatter(x='a', y='b',figsize=(10,3),s=25, c='green')
```

#### using a histogram plot to visualize column c

```python
# data index ['c'], calling histogram plot

df['c'].plot.hist()
```

```python
# using some arguments changing bins number

df['c'].plot.hist(bins=50)
```

#### try using plot style, and then calling a plot

```python
# using style.use method 

plt.style.use('ggplot')
```

```python
# after applying the style method we can see the plot for a similar column has changed.

df['c'].plot.hist(bins=50, alpha=0.5)
```

#### creating a boxplot comparing columns (a, b, c)

```python
# call box plot

df[['a','b','c']].plot.box()
```

#### creating KDE plot for column d

```python
df['d'].plot.kde()
```

```python
# using arguments on KDE plot

df['d'].plot.kde(lw='3', ls='--')
```

#### create an area plot of all columns, only the rows up to 50

first, we will use slicing to choose the rows we will work with using iloc method

```python
df.loc[0:50]

# or df.iloc[0:51]
```

```python
# calling areaplot

df.loc[0:50].plot.area()
```

```python
# use areaplot arguments 

df.loc[0:50].plot.area(alpha=0.6)
```

```python
# more arguments, putting the legend outside

fig = plt.figure()
df.loc[0:50].plot.area(alpha=0.6)
plt.legend(loc='center left', bbox_to_anchor=(1.0,0.5))
```
